# ✅ MCD_VAT
MakerDao transaction

    # MCD_VAT --> MakerDao data (DATE between 2019-11-01 to 2023-04-01)

    SELECT  block_timestamp,
            block_number,
            transaction_index,
            trace_address,
            transaction_hash,
            input,
            ARRAY_TO_STRING(
                ARRAY(
                    SELECT CHR(CAST(num AS INT64)) 
                    FROM UNNEST(SPLIT(trace_address, ',')) AS num
                    ),
                    ","
            ) as trace_addr_str
    FROM    `bigquery-public-data.crypto_ethereum.traces`
    WHERE   DATE(block_timestamp) >= "2023-02-01" 
            and DATE(block_timestamp) < "2023-04-01"
            and trace_address IS NOT NULL
            and call_type = "call"
            and to_address = LOWER("0x35D1b3F3D7966A1DFe207aa4514C12a259A0492B")
            and status = 1
    ORDER BY CAST(block_number as INT64), CAST(transaction_index as INT64), trace_addr_str

```python
class Vat(object):
  index = ['fn_name', 'ilk', 'usr', 'src', 'state']

  def __init__(self):
    pass

  def get_data(self, fn_name, ilk, usr, src=None):
    data = [fn_name, ilk, usr, src, None]
    return data

  def parse_init(self, fn_args):
    ilk = fn_args['ilk']
    data = ['init', ilk, None, None, None]
    return pd.Series(data=data, index=Vat.index)

  def parse_frob(self, fn_args):
    ilk = fn_args['i']
    u = fn_args['u']
    v = fn_args['v']
    data = self.get_data('frob', ilk, u, v)
    return pd.Series(data=data, index=Vat.index)

  def parse_grab(self, fn_args):
    ilk = fn_args['i']
    u = fn_args['u']
    data = self.get_data('grab', ilk, u)
    return pd.Series(data=data, index=Vat.index)

  def parse_slip(self, fn_args):
    ilk = fn_args['ilk']
    usr = fn_args['usr']
    wad = fn_args['wad']
    data = self.get_data('slip', ilk, usr)
    return pd.Series(data=data, index=Vat.index)

  def parse_fold(self, fn_args):
    ilk = fn_args['i']
    rate = fn_args['rate']
    data = ['fold', ilk, None, None, { "rate": rate }]
    return pd.Series(data=data, index=Vat.index)

  def parse_file(self, fn_args):
    ilk = fn_args.get('ilk', None)
    ilk = ilk if ilk else None
    data = fn_args['data']
    what = fn_args['what'].decode('utf-8')[:4]
    data = ['file', ilk, None, None, {"data": data, "what": what}]
    return pd.Series(data=data, index=Vat.index)

  def parse_fork(self, fn_args):
    ilk = fn_args['ilk']
    src = fn_args['src']
    dst = fn_args['dst']
    data = self.get_data('fork', ilk, dst, src)
    return pd.Series(data=data, index=Vat.index)

  def parse_flux(self, fn_args):
    ilk = fn_args['ilk']
    dst = fn_args['dst']
    src = fn_args['src']
    data = self.get_data('flux', ilk, dst, src)
    return pd.Series(data=data, index=Vat.index)
# -------------------------------------------------------------------------------------------------------------------
def decode_args(df, vat):
  errors = []

  def decode(tx_input):
    index = ['fn_name', 'ilk', 'usr', 'fn_args']

    fn_name = tx_input['fn_name']
    fn_args = tx_input['args']

    obj = fn_args.replace("b'", "'")
    obj = obj.replace("\\x00", "")
    obj = obj.replace("\'", "\"")
    res = json.loads(obj)

    timestamp = pd.Series([tx_input['timestamp']], index=['timestamp'])

    ilk = None
    src = None
    dst = None
    series = None

    dart = res.get('dart', None)
    dink = res.get('dink', None)
    wad = res.get('wad', None)

    ddw = pd.Series([dart, dink, wad], index=['dart', 'dink', 'wad'])

    try:
      if fn_name == 'init':
        series = vat.parse_init(res)

      if fn_name == 'slip':
        series = vat.parse_slip(res)

      if fn_name == 'grab':
        series = vat.parse_grab(res)

      if fn_name == 'frob':
        series = vat.parse_frob(res)

      if fn_name == 'fork':
        series = vat.parse_fork(res)

      if fn_name == 'flux':
        series = vat.parse_flux(res)

      if fn_name == 'fold':
        series = vat.parse_fold(res)

      if fn_name == 'file':
        series = vat.parse_file(res)
      
      series = series.append(timestamp)
      series = series.append(ddw)

      return series

    except Exception as e:
      print(e)
      print (fn_name)
      print (res)
      
      return None

  data = []
  for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    res = decode(row)
    if res is not None:
      data.append(res)

  return pd.DataFrame(data=data, columns=['fn_name', 'ilk', 'usr', 'src', 'state', 'timestamp', 'dart', 'dink', 'wad']), errors
```

```python
class Vat(object):
  index = ['fn_name', 'ilk', 'usr', 'src', 'state']

  def __init__(self):
    pass

  def get_data(self, fn_name, ilk, usr, src=None):
    data = [fn_name, ilk, usr, src, None]
    return data

  def parse_init(self, fn_args):
    ilk = fn_args['ilk']
    data = ['init', ilk, None, None, None]
    return pd.Series(data=data, index=Vat.index)

  def parse_frob(self, fn_args):
    ilk = fn_args['i']
    u = fn_args['u']
    v = fn_args['v']
    data = self.get_data('frob', ilk, u, v)
    return pd.Series(data=data, index=Vat.index)

  def parse_grab(self, fn_args):
    ilk = fn_args['i']
    u = fn_args['u']
    data = self.get_data('grab', ilk, u)
    return pd.Series(data=data, index=Vat.index)

  def parse_slip(self, fn_args):
    ilk = fn_args['ilk']
    usr = fn_args['usr']
    wad = fn_args['wad']
    data = self.get_data('slip', ilk, usr)
    return pd.Series(data=data, index=Vat.index)

  def parse_fold(self, fn_args):
    ilk = fn_args['i']
    rate = fn_args['rate']
    data = ['fold', ilk, None, None, { "rate": rate }]
    return pd.Series(data=data, index=Vat.index)

  def parse_file(self, fn_args):
    ilk = fn_args.get('ilk', None)
    ilk = ilk if ilk else None
    data = fn_args['data']
    what = fn_args['what'].decode('utf-8')[:4]
    data = ['file', ilk, None, None, {"data": data, "what": what}]
    return pd.Series(data=data, index=Vat.index)

  def parse_fork(self, fn_args):
    ilk = fn_args['ilk']
    src = fn_args['src']
    dst = fn_args['dst']
    data = self.get_data('fork', ilk, dst, src)
    return pd.Series(data=data, index=Vat.index)

  def parse_flux(self, fn_args):
    ilk = fn_args['ilk']
    dst = fn_args['dst']
    src = fn_args['src']
    data = self.get_data('flux', ilk, dst, src)
    return pd.Series(data=data, index=Vat.index)
```

```python
def decode_args(df, vat):
  errors = []

  def decode(tx_input):
    index = ['fn_name', 'ilk', 'usr', 'fn_args']

    fn_name = tx_input['fn_name']
    fn_args = tx_input['args']

    obj = fn_args.replace("b'", "'")
    obj = obj.replace("\\x00", "")
    obj = obj.replace("\'", "\"")
    res = json.loads(obj)

    timestamp = pd.Series([tx_input['timestamp']], index=['timestamp'])

    ilk = None
    src = None
    dst = None
    series = None

    dart = res.get('dart', None)
    dink = res.get('dink', None)
    wad = res.get('wad', None)

    ddw = pd.Series([dart, dink, wad], index=['dart', 'dink', 'wad'])

    try:
      if fn_name == 'init':
        series = vat.parse_init(res)

      if fn_name == 'slip':
        series = vat.parse_slip(res)

      if fn_name == 'grab':
        series = vat.parse_grab(res)

      if fn_name == 'frob':
        series = vat.parse_frob(res)

      if fn_name == 'fork':
        series = vat.parse_fork(res)

      if fn_name == 'flux':
        series = vat.parse_flux(res)

      if fn_name == 'fold':
        series = vat.parse_fold(res)

      if fn_name == 'file':
        series = vat.parse_file(res)
      
      series = series.append(timestamp)
      series = series.append(ddw)

      return series

    except Exception as e:
      print(e)
      print (fn_name)
      print (res)
      
      return None

  data = []
  for index, row in tqdm(df.iterrows(), total=df.shape[0]):
    res = decode(row)
    if res is not None:
      data.append(res)

  return pd.DataFrame(data=data, columns=['fn_name', 'ilk', 'usr', 'src', 'state', 'timestamp', 'dart', 'dink', 'wad']), errors
```

```python
def dai_eth(data):
  text_output = str(data)
  if len(text_output) < 66:
    dai_eth_rate = ""
  else:
    #print(int(text_output[:66],0))
    if int(text_output[:66],0) == 0:
      dai_eth_rate = ""
    else:
      dai_eth_rate = 1e18/(int(text_output[:66],0))
  return dai_eth_rate
```

```python
def eth_dai(data):
  text_output = str(data)
  if len(text_output) < 66:
    eth_dai_rate = ""
  else:
    #print(int(text_output[:66],0))
    if int(text_output[:66],0) == 0:
      eth_dai_rate = ""
    else:
      eth_dai_rate = (int(text_output[:66],0))/1e18
  return eth_dai_rate
```
